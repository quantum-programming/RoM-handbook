```python
import numpy as np
from exputils.actual_Amat import get_actual_Amat

A = get_actual_Amat(2)
A = A.tocoo()

val_to_idx = {}
val_to_idx[+1] = 0
val_to_idx[-1] = 1

print("\\foreach \\color/\\col/\\row in {")
for i, (val, row, col) in enumerate(zip(A.data, A.row, A.col)):
    idx = val_to_idx[val.real]
    print(f"c{idx}", col, -row, sep="/", end="")
    if i != A.nnz - 1:
        print(",")
    else:
        print("\n}{")
```
